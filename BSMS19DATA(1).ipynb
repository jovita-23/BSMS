```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
work=pd.read_csv("C:\\Users\\Jovita\\Documents\\BSMS 19 DATA.csv", encoding='ISO-8859–1')
work

```

```python
new_df=work['Educational goal of the student in the first year of study'].value_counts().rename_axis('edugoal_firstyear').reset_index(name='counts')
new_df
```

```python
your_labels=new_df.edugoal_firstyear
your_values=new_df.counts
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(your_values,labels=your_labels,autopct='%1.2f%%')
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Educational Goal in the First Year of Study",bbox={'facecolor':'0.8', 'pad':5})
plt.show()
```
